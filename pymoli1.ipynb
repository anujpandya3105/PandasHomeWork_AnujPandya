```python
group_SN = purchase_data_df.groupby(['SN','Age'])

purchase_data_d = pd.DataFrame(group_SN.head(len(group_SN)))

purchase_data_d = purchase_data_d.drop_duplicates('SN')
```

```python
#Age Demographics
print(purchase_data_d["Age"].max())
print(purchase_data_d["Age"].min())
```

```python
#Age Demographics
bins = [0,9,14,19,24,29,34, 39, 100]
group_name = ['<10','10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+' ]
```

```python
# Slice the age data and place it into bins
pd.cut(purchase_data_d["Age"],bins,labels=group_name)
```

```python
# Place the data series into a new column inside of the DataFrame
purchase_data_d["Age Group"] = pd.cut(purchase_data_d["Age"],bins,labels=group_name)
purchase_data_d.head()
```

```python
# Create a GroupBy object based upon "Age Group"
tot = len(purchase_data_d)
#print(tot)
purchase_group = purchase_data_d.groupby(["Age Group"])
#purchase_group = purchase_data_df.groupby(["Age Group", "SN"] )
#purchase_group.head(5)


                                                                                                                           
purchase_summary = pd.DataFrame( {"Total Count" :purchase_group["Age Group"].count(),
                                  "Percentage of Players" : round( purchase_group["Age Group"].count() *100 / tot ,2) })

purchase_summary
```

```python
purchase_count = purchase_group["Item ID"].count()
purchase_count
average_purchase_price = round(purchase_group["Price"].mean(),2)
average_purchase_price
total_purchase_value = round(purchase_group["Price"].sum())
total_purchase_value
normalized_total =total_purchase_value/ purchase_group["SN"].nunique()
normalized_total
```

```python
purchasing_analysis_summary = pd.DataFrame({"Purchase Count":purchase_count,
                                            "Average Purchase Price":average_purchase_price,
                                            "Total Purchase Value": total_purchase_value,
                                           "Normalized Totals":normalized_total })
purchasing_analysis_summary
```
